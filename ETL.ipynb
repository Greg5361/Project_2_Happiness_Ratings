```python
import pandas as pd
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import os
import requests
from bs4 import BeautifulSoup as bs
```

```python
df = pd.read_csv('Resources/world-happiness-report-2021.csv')
```

```python
df.head()
```

```python
df_filtered = df[["Country name", "Regional indicator", "Ladder score"]]
df_filtered
df_filtered.rename(columns={'Country name': 'Country', 'Regional indicator': 'Region', 'Ladder score': 'Rating'})

json = df_filtered.to_json(orient='records')
json

```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
browser.visit('https://en.wikipedia.org/wiki/Happiness')
description = browser.find_by_css('mw-body')
description
```
